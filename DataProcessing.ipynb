```python
df = pd.read_csv('orcid_name_id.csv')
df
```

```python
new_df = df.groupby('booktitle')['orcid'].apply(list)
```

```python
new_df = new_df.reset_index()
```

```python
new_df['len'] = new_df['orcid'].apply(len)
new_df = new_df.sort_values(by = 'len',ascending=False)
new_df
```

```python
def unique_list(lst):
    return list(set(lst))
```

```python
new_df['orcid_unique'] = new_df['orcid'].apply(unique_list)
new_df['unique_len'] = new_df['orcid_unique'].apply(len)
new_df = new_df.sort_values(by = 'unique_len',ascending=False)
new_df
```

```python
new_df.to_csv('groupby_orcid.csv')
```
